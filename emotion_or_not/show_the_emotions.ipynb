```python
used = set()
with open('possible_emotions.txt',encoding='utf-8') as f:
    with open('emotion_list.txt','w') as out:
        for line in f:
            if '(' not in line:
                continue
            par = str(line.split('(')[1].split(')')[0])
            par = par.lower()
            if 'to' in par or 'beat' in par:
                continue
            #print('to',par[:2])
            languages_list = ["Chinese", "Mandarin",  "English",  "Spanish",  "Arabic", "Bengali", "Hindi", "Russian",  "Portuguese",
                        "Japanese", "German",  "Korean",  "French",  "Turkish",  "Vietnamese",
                        "Cantonese",  "Tamil",  "Italian",  "Urdu",  "Polish","Ukrainian",  "Persian",
                        "Xiang",  "Malayalam",  "Panjabi", "Romanian",  "Azerbaijani",  "Burmese",
                        "Thai", "Dutch","Yiddish"]
            for lang in languages_list:
                if lang.lower() in par:
                    par = ""
                    break
            if par == '':
                continue
            if par in used:
                continue
            used.add(par)
            print('=',par,'=')
            out.write(par+'\n')
```
